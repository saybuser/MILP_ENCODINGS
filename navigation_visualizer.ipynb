```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readSolutionFile(directory):
    solnFile = open(directory,"r")
    solutions = solnFile.read().splitlines()
    np = []
    
    for index, solution in enumerate(solutions):
        if index > 1 and (solution != ";" and solution != ""):
            step, dim_x, dim_y = solution.split()
            np.append([dim_x] + [dim_y])
            
    solndf = pd.DataFrame(np)
    solndf.columns = ['x','y']
    
    return solndf          
```

```python
solndf = readSolutionFile("./soln_location.txt")
print(len(solndf)-1)
solndf
```

```python
def plotMaze(dataframe):
    fig = plt.figure(figsize=(5, 5), dpi=100)
    
    axis = fig.add_subplot(111)    
    axis.set_xlim([0, 10])
    axis.set_ylim([0, 10])
    
    #plot the path of the agent
    for step in range(len(dataframe)-1):
        plt.plot([dataframe['x'].iloc[step],dataframe['x'].iloc[step+1]], [dataframe['y'].iloc[step],dataframe['y'].iloc[step+1]],color='red',lw=2.0)
    
    #plot the goal region
    x = np.arange(7, 9, 0.01)
    axis.fill_between(x, 7, 9, facecolor='gold')
    
    #plot the mud region
    x = np.arange(4, 6, 0.01)
    axis.fill_between(x, 4, 6, facecolor='indianred')
    
    #plot the wind region
    for x in range(10+1):
        for y in range(5+1):
            axis.arrow(x,y,-0.5,0, color='blue', head_width=0.1)
    
    #set axis labels
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    
    #set legend
    plt.plot([0,0], [0,0], color='red',lw=1.5,label="Agent")
    plt.plot([0,0], [0,0], color='indianred',lw=1.5,label="Mud")
    plt.plot([0,0], [0,0], color='blue',lw=1.5,label="Wind")
    plt.plot([0,0], [0,0], color='gold',lw=1.5,label="Goal")
    legend = plt.legend(loc='lower right',prop={'size':10})
    
    plt.savefig('navigation.eps',format='eps')
```

```python
plotMaze(solndf)
```
